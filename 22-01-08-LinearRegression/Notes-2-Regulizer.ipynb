```python
import numpy as np
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
x = np.linspace(-5, 5, 100)
y = 2*x + 4
```

```python
plt.plot(x, y)
```

```python
X = x.reshape(-1, 1)
```

```python
from sklearn.linear_model import LinearRegression

lin = LinearRegression()
lin.fit(X, y)
```

```python
lin.coef_ , lin.intercept_
```

```python
plt.plot(x, y, label = 'actual')
plt.plot(x, lin.predict(X), label = 'predicted')

plt.legend()
```

```python
x = np.linspace(-5, 5, 100)
y = np.linspace(-4, 10, 100)
z = x*3 + y*7 + 10 + np.random.rand(x.size)*10

fig = px.scatter_3d(x = x, y = y, z = z)
fig.show()
```

```python
X = np.hstack([x.reshape(-1, 1), y.reshape(-1, 1)])
```

```python
lin.fit(X, z)
```

```python
lin.coef_, lin.intercept_
```

```python
import plotly.graph_objs as go

fig = go.Figure()

fig.add_trace(go.Scatter3d(x = x, y = y, z = z, name = 'actual'))
fig.add_trace(go.Scatter3d(x = x, y = y, z = lin.predict(X), name = 'predict'))
```
