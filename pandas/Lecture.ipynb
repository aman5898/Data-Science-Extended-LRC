# DataSet

```python
df = pd.read_csv('https://raw.githubusercontent.com/neylsoncrepalde/projeto_eda_covid/master/covid_19_data.csv')
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df.head()
```

```python
df = df.rename(columns={'ObservationDate': 'Date', 'Province/State': 'State', 'Country/Region': 'Country', })
```

```python
df.head()
```

```python
df.drop(columns=['SNo', 'Last Update'], axis = 1, inplace = True)
```

```python
df.head()
```

```python
df.info()
```

```python
df['Date'] = pd.to_datetime( df['Date'] )
```

```python
df.info()
```

```python
df.head()
```

```python
df.groupby('Country').sum().reset_index()
```

```python
df.groupby(['Country', 'Date']).sum().reset_index()
```
